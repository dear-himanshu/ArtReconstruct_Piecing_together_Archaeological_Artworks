```python
import cv2
import numpy as np

def remove_pink_background(image_path):
    # Read the image
    image = cv2.imread(image_path)
    
    # Convert the image from BGR to HSV color space
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    
    # Define lower and upper bounds for pink color in HSV
    lower_pink = np.array([150, 50, 50])
    upper_pink = np.array([180, 255, 255])
    
    # Create a binary mask by thresholding the image based on the pink color bounds
    pink_mask = cv2.inRange(hsv_image, lower_pink, upper_pink)
    
    # Invert the mask
    pink_mask_inverse = cv2.bitwise_not(pink_mask)
    
    # Apply the mask to the original image to remove the pink background
    result_image = cv2.bitwise_and(image, image, mask=pink_mask_inverse)
    
    return result_image

def main():
    # Provide the path to the image
    image_path = "fragment_0005.png"
    
    # Remove the pink background
    result_image = remove_pink_background(image_path)
    
    # Display the result
    cv2.imshow("Result Image", result_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```

```python
# Python code to find the co-ordinates of 
# the contours detected in an image. 
import numpy as np 
import cv2 

# Reading image 
font = cv2.FONT_HERSHEY_COMPLEX 
img2 = cv2.imread('cont.png', cv2.IMREAD_COLOR) 

# Reading same image in another 
# variable and converting to gray scale. 
img = cv2.imread('cont.png', cv2.IMREAD_GRAYSCALE) 

# Converting image to a binary image 
# ( black and white only image). 
_, threshold = cv2.threshold(img, 110, 255, cv2.THRESH_BINARY) 

# Detecting contours in image. 
contours, _= cv2.findContours(threshold, cv2.RETR_TREE, 
							cv2.CHAIN_APPROX_NONE) 

# Going through every contours found in the image. 
for cnt in contours : 

	approx = cv2.approxPolyDP(cnt, 0.009 * cv2.arcLength(cnt, True), True) 

	# draws boundary of contours. 
	cv2.drawContours(img2, [approx], 0, (0, 0, 255), 5) 

	# Used to flatted the array containing 
	# the co-ordinates of the vertices. 
	n = approx.ravel() 
	i = 0

	for j in n : 
		if(i % 2 == 0): 
			x = n[i] 
			y = n[i + 1] 

			# String containing the co-ordinates. 
			string = str(x) + " " + str(y) 

			if(i == 0): 
				# text on topmost co-ordinate. 
				cv2.putText(img2, "Arrow tip", (x, y), 
								font, 0.5, (255, 0, 0)) 
			else: 
				# text on remaining co-ordinates. 
				cv2.putText(img2, string, (x, y), 
						font, 0.5, (0, 255, 0)) 
		i = i + 1

# Showing the final image. 
cv2.imshow('image2', img2) 

# Exiting the window if 'q' is pressed on the keyboard. 
if cv2.waitKey(0) & 0xFF == ord('q'): 
	cv2.destroyAllWindows() 
```

```python
#code to extract contour points 
import cv2

# Read the image
image = cv2.imread("cont.png")

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Find contours in the image
contours, _ = cv2.findContours(gray, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

# Assuming you have only one contour (change the index if you have multiple)
contour = contours[0]

# Get the coordinates of the contour points
coordinates_1 = []
for point in contour:
    x, y = point[0]
    coordinates_1.append((x, y))

print("Coordinates of contour points:")
print(coordinates_1)
print(len(coordinates_1))
```

```python
# directly use this to find contour coordinates and plot 
import matplotlib.pyplot as plt

# plt.plot(coordinates)
import cv2
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread("cont.png")

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Find contours in the image
contours, _ = cv2.findContours(gray, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

# Assuming you have only one contour (change the index if you have multiple)
contour = contours[0]

# Get the coordinates of the contour points
x_coords = [point[0][0] for point in contour]
y_coords = [point[0][1] for point in contour]

# Plot the contour
plt.figure(figsize=(6, 6))
plt.plot(x_coords, y_coords, '-r', linewidth=0.5)
print(len(x_coords))
plt.gca().invert_yaxis()  # Invert y-axis to match image coordinates
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Contour Coordinates')
plt.grid()
plt.show()
```
